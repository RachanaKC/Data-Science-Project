```python
from bs4 import BeautifulSoup
import requests
url =BeautifulSoup('https://www.sharesansar.com/live-trading/','html.parser')
page = requests.get(url)

```

```python
soup = BeautifulSoup(page.text, "html")
soup = soup.table
print(soup)
```

```python
data = []
tags = soup.find_all('tr')
for i in tags:
  y = i.text.split('\n')[1:]
  if y[0] !=" ":
    data.append(y)

data
```

```python
data2 = []
for sub_list in data:
    new_sub_list = list(filter(None, sub_list))
    data2.append(new_sub_list)

print(data2)
```

```python
import csv
file = open('share.csv','w')
x = csv.writer(file)
x.writerows(data2)
file.close()
```

```python
import pandas as pd
df = pd.read_csv('share.csv',encoding = 'latin1')
df.dropna()
df
```

```python
df = df[0:10]
df['LTP'] = df['LTP'].str.replace(",","")
df['LTP'] = df['LTP'].str.replace(" ","")
df['LTP']= df['LTP'].astype('float').astype('int')

df['Open'] = df['Open'].str.replace(",","")
df['Open'] = df['Open'].str.replace(" ","")
df['Open']= df['Open'].astype('float').astype('int')

df['High'] = df['High'].str.replace(",","")
df['High'] = df['High'].str.replace(" ","")
df['High']= df['High'].astype('float').astype('int')

df['Low'] = df['Low'].str.replace(",","")
df['Low'] = df['Low'].str.replace(" ","")
df['Low']= df['Low'].astype('float').astype('int')

df['Volume'] = df['Volume'].str.replace(",","")
df['Volume'] = df['Volume'].str.replace(" ","")
df['Volume']= df['Volume'].astype('float').astype('int')

df['Prev. Close'] = df['Prev. Close'].str.replace(",","")
df['Prev. Close'] = df['Prev. Close'].str.replace(" ","")
df['Prev. Close']= df['Prev. Close'].astype('float').astype('int')
df.info()
```
